```python
from plotnine import ggplot, aes, labs, geom_point
from plotnine.data import mtcars

p = ggplot(mtcars, aes("mpg", "wt", colour = "cyl", size="gear")) + geom_point()
```

### Adding a title and axis labels

```python
# Gallery, themesandlabels
p + labs(
    title = "The title",
    subtitle = "The subtitle",
    x = "New x label",
    y = "New y label",
    caption = "(A nice caption)"
)
```

### Labelling scales

```python
# Gallery, themesandlabels
p + labs(colour = "Cylinders", size = "Gears")
```
