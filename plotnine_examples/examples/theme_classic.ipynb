```python
from plotnine import ggplot, aes, labs, theme_classic, theme_bw, theme_void, geom_point
from plotnine.data import mtcars

p = (
    ggplot(mtcars, aes(x="wt", y="mpg", colour="factor(gear)"))
    + geom_point()
    + labs(
        title = "Fuel economy declines as weight increases",
        subtitle = "(1973-74)",
        caption = "Data from the 1974 Motor Trend US magazine.",
        x = "Weight (1000 lbs)",
        y = "Fuel economy (mpg)",
        colour = "Gears"
    )
)
```

### theme_classic()

```python
# Gallery, themesandlabels
p + theme_classic()
```
