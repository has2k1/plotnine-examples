```python
from plotnine import (
    ggplot,
    aes,
    theme_matplotlib,
    theme_set,
    geom_tile,
    scale_fill_continuous,
)

from plotnine.data import faithfuld

# Set default theme for all the plots
theme_set(theme_matplotlib())
```

### Default scale fill

Without specifically calling `scale_fill_continuous()`, a default scale is applied to `geom_tile()`.

```python
v = ggplot(faithfuld, aes("waiting", "eruptions", fill="density")) + geom_tile()
v
```

Notice that adding `scale_fill_continuous()` will not change the default plot.

```python
v + scale_fill_continuous()
```

The scale aesthetics can be edited to other color maps. Use the `cmap_name` argument to specify an alternative palette. 

```python
# find color maps at 
# https://matplotlib.org/stable/users/explain/colors/colormaps.html#classes-of-colormaps
v + scale_fill_continuous(cmap_name="Purples")
```
